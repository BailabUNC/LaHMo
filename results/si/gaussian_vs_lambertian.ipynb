```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

mpl.rcParams['font.family'] = 'arial'
mpl.rcParams['font.size'] = 14

def gaussian_2d(x, y, mu_x, mu_y, sigma_x, sigma_y):
    return np.exp(-((x - mu_x)**2 / (2 * sigma_x**2) + (y - mu_y)**2 / (2 * sigma_y**2)))

def lambertian(x, y):
    return np.cos(np.arctan2(np.sqrt(x**2 + y**2), 1))

# Generate a grid of points
x = np.linspace(-2, 2, 500)
y = np.linspace(-2, 2, 500)
x, y = np.meshgrid(x, y)

# Define the parameters for the Gaussian distribution
mu_x = 0
mu_y = 0
sigma_x = 1
sigma_y = 1

# Compute the Gaussian and Lambertian distributions
gaussian_distribution = gaussian_2d(x, y, mu_x, mu_y, sigma_x, sigma_y)
lambertian_distribution = lambertian(x, y)

# Create a figure and subplots for 3D plotting
fig = plt.figure(figsize=(12, 5), constrained_layout=True)
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')

# Plot the Gaussian distribution
ax1.plot_surface(x, y, gaussian_distribution, cmap='viridis', linewidth=0, antialiased=False,
                 rstride=1, cstride=1)
ax1.set_title('Two-Element Gaussian Distribution')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Intensity')

# Plot the Lambertian distribution
ax2.plot_surface(x, y, lambertian_distribution, cmap='viridis', linewidth=0, antialiased=False,
                 rstride=1, cstride=1)
ax2.set_title('Lambertian Distribution')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Intensity')

# plt.subplots_adjust(left=0.05, right=0.95, top=0.9, bottom=0.1, wspace=0.1, hspace=0.3)
# plt.tight_layout()
ax1.set_box_aspect(None, zoom=0.85)
ax2.set_box_aspect(None, zoom=0.85)
plt.savefig('gaussian_vs_lambertian.png', dpi=600, bbox_inches='tight')
plt.show()
```
